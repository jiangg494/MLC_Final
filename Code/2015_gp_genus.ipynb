```python
import pandas as pd
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel
from matplotlib import pyplot as plt
from pgmpy.estimators import HillClimbSearch, BicScore
import matplotlib.pylab as plt
```

```python
df_new.columns
```

```python
X_obs = df_new[['latitude', 'longitude']]
y_obs = df_new[['genus_Acer', 'genus_Fraxinus', 'genus_Ginkgo',
       'genus_Gleditsia', 'genus_Koelreuteria', 'genus_Liquidambar',
       'genus_Platanus', 'genus_Prunus', 'genus_Pyrus', 'genus_Quercus',
       'genus_Styphnolobium', 'genus_Syringa', 'genus_Tilia', 'genus_Ulmus',
       'genus_Zelkova']]
```

```python
kernel = 1.* RBF(length_scale=1., length_scale_bounds=(1e-3, 1e3)) \
    + WhiteKernel(noise_level=1., noise_level_bounds=(1e-10, 1e10))

gp = GaussianProcessRegressor(kernel=kernel,random_state=1)

gp.fit(X_obs, y_obs)

y_mesh_pred, sigma = gp.predict(X_obs, return_std=True)

print(gp.kernel_)
```
